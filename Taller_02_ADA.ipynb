## Uso de statsmodels para an치lisis de resultados

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
```

## Carga de los archivos


```python
df_results = pd.read_csv('timestamps.res', sep=',', header=None, names=['n','Iterativo', 'Recursivo' ])
```

```python
df_results.head()
```

```python
df_results.plot(y=['Iterativo','Recursivo'], x='n')
```

```python
mod = smf.ols('Iterativo ~ np.power(n,2)', data=df_results)
res = mod.fit()
print(res.summary())
```

```python
res.params
```

```python
sm.graphics.plot_fit(res,"np.power(n, 2)")
```

### Recursivo

```python
mod = smf.ols(formula='Recursivo ~ np.power(n,2)', data=df_results)
```

```python
res = mod.fit()
```

```python
print(res.summary())
```

```python
res.params
```

```python
sm.graphics.plot_fit(res,"np.power(n, 2)")
```
